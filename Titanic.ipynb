```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import random
```

```python
# 원자료 GET
f = open('train.csv','r')
df = pd.read_csv("train.csv")
df.index.name = "Num"
df.columns.name = "info"
```

```python
# 랜덤으로 모집단 추출
ran_list = []
ran_num=0
for i in range(30):
    while ran_num in ran_list:
        ran_num = random.randint(0,890)
    ran_list.append(ran_num)
ran_list.sort()
pop_df = df.loc[ran_list]
pop_df
```

```python
# 전체 데이터 셋에 대한 성별 승객 수 시각화
df_age = df['Age']
age_cnt = df_age.value_counts().sort_index() # 각 나이 빈도 수 추출 후 오름차순으로 정렬
age_cnt.plot(kind = 'bar', title = 'age counter') # 내림차순 정렬 후 빈도 수 시각화
plt.rcParams["figure.figsize"] = (30, 20)
plt.show()
```
